```python
top_countries = df["country"].value_counts().head(5)
top_countries
```

```python
highest_adr_person = df.loc[df["adr"].idxmax(),"name"]
highest_adr_person
```

```python
avg_nights = (df["stays_in_week_nights"] + df["stays_in_weekend_nights"]).mean()
avg_nights
```

```python
df["total_cost"] = df["adr"]* (df["stays_in_week_nights"] + df["stays_in_weekend_nights"])
average_totalcost = df["total_cost"].mean()
average_totalcost
```

```python
df['last_name'] = df['name'].apply(lambda x: x.split()[-1])
top_last_names = df['last_name'].value_counts().head(5)
top_last_names
```

```python
most_children_babies = df.groupby('name')[['children', 'babies']].sum().sum(axis=1).idxmax()
most_children_babies
```

```python
df['area_code'] = df['phone-number'].apply(lambda x: x.split('-')[0])
top_area_codes = df['area_code'].value_counts().head(3)
top_area_codes
```

```python
arrivals_1_15 = df[df['arrival_date_day_of_month'].between(1, 15)].shape[0]
arrivals_1_15
```
